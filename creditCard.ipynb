this dataset is highly unbalance

0 --> formal transaction

1-->fraud transaction

```python
#separating the data to analyse
legt=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
legt.shape , fraud.shape
```

```python
legt.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values of the two classes
credit_card_data.groupby('Class').mean()
```

under sampling

build a sample dataset contain similar distribution of formal and fraud transaction

number of rows of fraud transaction = 492

```python
legt_sample=legt.sample(n=492)
```

concatinating the two dataframe

```python
credit_data=pd.concat([legt_sample,fraud],axis=0)
credit_data.shape
```

```python
credit_data['Class'].value_counts()
```

```python
credit_data.groupby('Class').mean()
```

```python
x=credit_data.drop(columns='Class',axis=1)
y=credit_data['Class']
```

```python
print(x,y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2,stratify=y)
```

```python
x.shape,x_train.shape,x_test.shape
```

fit the model

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_prediction=model.predict(x_train)
x_train_accuracy=accuracy_score(x_train_prediction,y_train)
print("the train data accuracy: ",x_train_accuracy)
```

```python
x_test_prediction=model.predict(x_test)
x_test_accuracy=accuracy_score(x_test_prediction,y_test)
print("the test data accuracy: ",x_test_accuracy)
```
